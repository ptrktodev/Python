üî± **Hierarquia de √≠ndices e √≠ndices M√∫ltiplos**

```python
outside = ['H', 'H','M', 'M']
inside = [1, 2, 1, 2]

hierq_ind = list(zip(outside, inside))

print(hierq_ind)
```

```python
# transformar em um objeto MultiIndex. Usado como um √≠ndice em um DataFrame do pandas para criar uma estrutura de √≠ndice hier√°rquico.
hierq_ind = pd.MultiIndex.from_tuples(hierq_ind) 
```

```python
df3 = pd.DataFrame(randn(4, 2), index=hierq_ind, columns='A B'.split())
df3
```

```python
# para renomear esse indices novos criados:
df3.index.names = ['G√™nero', 'N√∫mero']
df3
```

```python
# BUSCAS E SELE√á√ÉO:

df3.loc['M'].loc[:, ['A']]
df3.loc['M'].loc[[2]]
df3.loc['M']
```

üß∂

üß† **METHOD: GROUP BY**

Usado para agrupar dados em um DataFrame com base em uma ou mais colunas.
> Obs: em [sum(), mean()...] defina numeric_only=True, as fun√ß√µes aplicar√£o a opera√ß√£o apenas √†s colunas que cont√™m dados num√©ricos.

```python
vendas_funcionario = {
    'Classe': ['Pleno', 'Senior', 'Junior', 'Senior', 'Pleno', 'Junior', 'Pleno', 'Senior', 'Junior', 'Pleno'],
    'Funcion√°rios': ['Beto', 'Larissa', 'Anderson', 'Lucas', 'Mariana', 'Jo√£o', 'Pedro', 'Isabela', 'Maria', 'Ricardo'],
    'Vendas': [1000, 1500, 800, 1200, 2000, 750, 1800, 1600, 850, 1700]
}

dfData = pd.DataFrame(vendas_funcionario)
dfData
```

```python
# Qual foi o total de vendas por classe:

group = dfData.groupby('Classe').sum(numeric_only=True) 
group
```

‚úàÔ∏è **METHOD: CONCAT**

Serve para empilhar objetos do tipo pandas, como DataFrames ou Series, ao longo de um eixo espec√≠fico. 

> 1. Se compartilham das Mesmas Colunas -> axis=0

> 2. Se compartilham das Mesmas Linhas -> axis=1

```python
df_1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3'],
                        'C': ['C0', 'C1', 'C2', 'C3'],
                        'D': ['D0', 'D1', 'D2', 'D3']},
                        index=[0, 1, 2, 3])

df_2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                        'B': ['B4', 'B5', 'B6', 'B7'],
                        'C': ['C4', 'C5', 'C6', 'C7'],
                        'D': ['D4', 'D5', 'D6', 'D7']},
                         index=[4, 5, 6, 7])
df_3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                        'B': ['B8', 'B9', 'B10', 'B11'],
                        'C': ['C8', 'C9', 'C10', 'C11'],
                        'D': ['D8', 'D9', 'D10', 'D11']},
                        index=[8, 9, 10, 11])
```

```python
pd.concat([df_1, df_2, df_3], axis=0)
```

ü§∏‚Äç‚ôÇÔ∏è **METHOD: MERGE**

Serve para combinar dois ou mais DataFrames em um √∫nico DataFrame com base em uma ou mais colunas **compartilhadas**.

```python
esquerda = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
direita = pd.DataFrame({'Key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})   
```

```python
# [on='Key'] especifica qual coluna (ou colunas) deve ser usada como chave para realizar a jun√ß√£o (merge) 
pd.merge(esquerda, direita, on='Key') 
```

> **OUTRO EXEMPLO ABAIXO:**

```python
tri1 = {
    'Classe': ['Pleno', 'Senior', 'Junior', 'Senior', 'Pleno', 'Junior', 'Pleno', 'Senior', 'Junior', 'Pleno'],
    'Funcion√°rios': ['Beto', 'Larissa', 'Anderson', 'Lucas', 'Mariana', 'Jo√£o', 'Pedro', 'Isabela', 'Maria', 'Ricardo'],
    'Vendas 1¬∫': [1000, 1500, 800, 1200, 2000, 750, 1800, 1600, 850, 1700]
}

tri2 = {
    'Classe': ['Pleno', 'Senior', 'Junior', 'Senior', 'Pleno', 'Junior', 'Pleno', 'Senior', 'Junior', 'Pleno'],
    'Funcion√°rios': ['Beto', 'Larissa', 'Anderson', 'Lucas', 'Mariana', 'Jo√£o', 'Pedro', 'Isabela', 'Maria', 'Ricardo'],
    'Vendas 2¬∫': [2345, 2346, 2341, 5673, 2356, 457320, 334, 4572, 2547, 347]
}

tri3 = {
    'Classe': ['Pleno', 'Senior', 'Junior', 'Senior', 'Pleno', 'Junior', 'Pleno', 'Senior', 'Junior', 'Pleno'],
    'Funcion√°rios': ['Beto', 'Larissa', 'Anderson', 'Lucas', 'Mariana', 'Jo√£o', 'Pedro', 'Isabela', 'Maria', 'Ricardo'],
    'Vendas 3¬∫': [1245, 8562, 2341, 224, 534, 6792, 461, 36, 561, 23541]
}


tri1 = pd.DataFrame(tri1)
tri2 = pd.DataFrame(tri2)
tri3 = pd.DataFrame(tri3)
```

```python
total = tri1.merge(tri2, on=['Classe', 'Funcion√°rios'])
total = total.merge(tri3, on=['Classe', 'Funcion√°rios'])
total['Soma Vendas'] = total['Vendas 1¬∫'] + total['Vendas 2¬∫'] + total['Vendas 3¬∫']
```

```python
total
```

üí§ **APLICANDO FUN√á√ïES:**
> O m√©todo apply() em pandas √© usado para aplicar uma fun√ß√£o a elementos de um objeto DataFrame ou Series. Retorna uma Series.

> total['Soma Vendas x2'] = total['Soma Vendas'].apply(lambda x: x * 2)

```python
def mult(x):
    return x * 2
```

```python
# Normal
total['Soma Vendas x2'] = total['Soma Vendas'].apply(mult)
```

```python
# com Lambda
total['Soma Vendas x3'] = total['Soma Vendas'].apply(lambda x: x * 3)
```
